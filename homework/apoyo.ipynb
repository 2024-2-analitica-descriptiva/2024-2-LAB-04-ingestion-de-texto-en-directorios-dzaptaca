````python
"""
    La información requerida para este laboratio esta almacenada en el
    archivo "files/input.zip" ubicado en la carpeta raíz.
    Descomprima este archivo.

    Como resultado se creara la carpeta "input" en la raiz del
    repositorio, la cual contiene la siguiente estructura de archivos:


    ```
    train/
        negative/
            0000.txt
            0001.txt
            ...
        positive/
            0000.txt
            0001.txt
            ...
        neutral/
            0000.txt
            0001.txt
            ...
    test/
        negative/
            0000.txt
            0001.txt
            ...
        positive/
            0000.txt
            0001.txt
            ...
        neutral/
            0000.txt
            0001.txt
            ...
"""
import os
import zipfile

# Ruta del archivo ZIP
zip_file_path = "../files/input.zip"

# Ruta donde se descomprimirá el contenido
destination_folder = "../input"

# Verificar si el archivo ZIP existe
if os.path.exists(zip_file_path):
    # Abrir y extraer el archivo ZIP
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(destination_folder)
        print(f"Archivo descomprimido en: {destination_folder}")

    # Listar la estructura de carpetas y archivos
    for root, dirs, files in os.walk(destination_folder):
        level = root.replace(destination_folder, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(root)}/")
        sub_indent = ' ' * 4 * (level + 1)
        for file in files:
            print(f"{sub_indent}{file}")
else:
    print("El archivo ZIP no existe en la ruta especificada.")
````

````python
import os
import pandas as pd

def create_csv_from_text(input_dir, output_path):
    data = []

    for sentiment in os.listdir(input_dir):
        sentiment_path = os.path.join(input_dir, sentiment)
        if os.path.isdir(sentiment_path):
            for file_name in os.listdir(sentiment_path):
                file_path = os.path.join(sentiment_path, file_name)
                with open(file_path, 'r', encoding='utf-8') as file:
                    phrase = file.read().strip()
                    data.append({'phrase': phrase, 'sentiment': sentiment})

    df = pd.DataFrame(data)
    df.to_csv(output_path, index=False)

# Definimos las rutas
input_base_path = '../input'
output_base_path = '../output'

# Crear la carpeta de salida si no existe
os.makedirs(output_base_path, exist_ok=True)

# # Procesar los datasets de train y test
# create_csv_from_text('../input/test', '../output/train_dataset.csv')
# create_csv_from_text('../input/train', '../output/test_dataset.csv')

# Procesar los datasets de train y test
create_csv_from_text(os.path.join(input_base_path, 'train'), os.path.join(output_base_path, 'train_dataset.csv'))
create_csv_from_text(os.path.join(input_base_path, 'test'), os.path.join(output_base_path, 'test_dataset.csv'))

print("Archivos CSV generados exitosamente en la carpeta 'output'.")
````
